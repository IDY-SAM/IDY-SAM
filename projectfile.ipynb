```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
```

```python
val1 =pd.read_csv('./Divvy_Trips_2019_Q2.csv', skipinitialspace=True)
val2 =pd.read_csv('./Divvy_Trips_2019_Q3.csv', skipinitialspace=True)
val3 =pd.read_csv('./Divvy_Trips_2019_Q4.csv', skipinitialspace=True)
val4 =pd.read_csv('./Divvy_Trips_2020_Q1.csv', skipinitialspace=True)
```

```python
#renaming columns in val1
val1.rename(columns={"01 - Rental Details Rental ID":"ride_id"}, inplace=True)
val1.rename(columns={"01 - Rental Details Bike ID":"rideable_type"}, inplace=True)
val1.rename(columns={"01 - Rental Details Local Start Time":"started_at"}, inplace=True)
val1.rename(columns={"01 - Rental Details Local End Time":"ended_at"}, inplace=True)
val1.rename(columns={"03 - Rental Start Station Name":"start_station_name"}, inplace=True)
val1.rename(columns={"03 - Rental Start Station ID":"start_station_id"}, inplace=True)
val1.rename(columns={"02 - Rental End Station Name":"end_station_name"}, inplace=True)
val1.rename(columns={"02 - Rental End Station ID":"end_station_id"}, inplace=True)
val1.rename(columns={"User Type":"member_casual"}, inplace=True)
```

```python
#renaming columns in val2
val2.rename(columns={"trip_id":"ride_id"}, inplace=True)
val2.rename(columns={"bikeid":"rideable_type"}, inplace=True)
val2.rename(columns={"start_time":"started_at"}, inplace=True)
val2.rename(columns={"end_time":"ended_at"}, inplace=True)
val2.rename(columns={"from_station_name":"start_station_name"}, inplace=True)
val2.rename(columns={"from_station_id":"start_station_id"}, inplace=True)
val2.rename(columns={"to_station_name":"end_station_name"}, inplace=True)
val2.rename(columns={"to_station_id":"end_station_id"}, inplace=True)
val2.rename(columns={"usertype":"member_casual"}, inplace=True)
```

```python
#renaming columns in val3
val3.rename(columns={"trip_id":"ride_id"}, inplace=True)
val3.rename(columns={"bikeid":"rideable_type"}, inplace=True)
val3.rename(columns={"start_time":"started_at"}, inplace=True)
val3.rename(columns={"end_time":"ended_at"}, inplace=True)
val3.rename(columns={"from_station_name":"start_station_name"}, inplace=True)
val3.rename(columns={"from_station_id":"start_station_id"}, inplace=True)
val3.rename(columns={"to_station_name":"end_station_name"}, inplace=True)
val3.rename(columns={"to_station_id":"end_station_id"}, inplace=True)
val3.rename(columns={"usertype":"member_casual"}, inplace=True)
```

```python
#changing datatype in val1, val2, and val3 from integer to string to aline with the val4
val1["ride_id"]=val1["ride_id"].astype(str)
val1["rideable_type"]=val1["rideable_type"].astype(str)
val2["ride_id"]=val2["ride_id"].astype(str)
val2["rideable_type"]=val2["rideable_type"].astype(str)
val3["ride_id"]=val3["ride_id"].astype(str)
val3["rideable_type"]=val3["rideable_type"].astype(str)
val3.info()
```

```python
#joining the datasets together
#when combining datasets the columns must be the same number
join=pd.concat([val1,val2,val3,val4])
join
```

```python
join.info()
```

```python
#dropping columns
table= join.drop(columns=['01 - Rental Details Duration In Seconds Uncapped','Member Gender','05 - Member Details Member Birthday Year','tripduration','gender','birthyear','start_lat', 'start_lng','end_lat','end_lng'])
```

```python
#changing casual to customer and Subscriber to Member 
a = table['member_casual']=='casual'
table.loc[a,'member_casual']='Customer'
b = table['member_casual']=='member'
table.loc[b,'member_casual']= 'Subscriber'
table
```

```python
table.info()
```

```python
#changing a column datatype to datetime#
table['started_at']=pd.to_datetime(table['started_at'])
table['ended_at']=pd.to_datetime(table['ended_at'])
table
```

```python
#changing starting year to month_name
table['Year of Starting']=table['started_at'].dt.year
table['Month of starting']=table['started_at'].dt.month_name()
table['Day of starting']=table['started_at'].dt.day_name()
table['Year of ending']=table['ended_at'].dt.year
table['Month of ending']=table['ended_at'].dt.month_name()
table['Day of ending']=table['ended_at'].dt.day_name()
```

ANALYSING THE DATA

```python
#1. HOW MANY CUSTOMERS AND SUBSCRIBERS ARE THERE IN THE BIKERS COMPANY
ca= (table.groupby(["member_casual"]))
ca1=ca["member_casual"].value_counts()
ca1
```

```python
ab = ca1.to_list()
ab
```

```python
style.use('ggplot')
slices=ab
names=['Customers','Subscribers']
col=['#0000FF','#FFFF00']
plt.pie(slices,labels=names,colors=col,startangle=140,shadow=True,autopct="%1.1f %%")
# plt.plot([],[],color="#0000FF",label="Customer")
# plt.plot([],[],color="#FFFF00",label="Subscribers")
plt.title('TOTAL NUMBER OF CUSTOMERS AND SUBCRIBERS IN THE COMPANY')
#plt.legend()
plt.show()
```

```python
#2. WHAT STATION IS USED THE MOST AS AN END STATION
st=table.groupby(['end_station_name'])
st1=st['end_station_name'].value_counts()
#coverting the values to a dataframe
top_station= pd.DataFrame(st1)
#sorting the values to get the highest 10 accoding to the count
top_station.sort_values(by='count',ascending=False,inplace=True)
#sorting the first five values in the dataset for visualization
five=top_station.head()
#to get the highest 5 station names
five1=five['count'].index.tolist()
#to get the number of times used by the end station
five2=five['count'].tolist()
five1
```

```python
style.use('ggplot')
g=five1
h=five2
plt.bar(g,h,color='#0000FF',align='center')
plt.xticks(rotation=45,ha='right')
plt.title('THE MOST FREQUENTLY USED END STATION')
plt.ylabel('COUNT')
plt.xlabel('END STATIONS')
plt.show()
```

```python
#3. WHAT STATION IS USED THE MOST AS THE START STATION
x=table.groupby(['start_station_name'])
x1=x['start_station_name'].value_counts()
#converting the values to a DataFrame
top_start_stations=pd.DataFrame(x1)
#sorting the values to get the highest 10 according to the count
top_start_stations.sort_values(by='count', ascending=False,inplace=True)
#sorting the first five in a variables for visualization
x2=top_start_stations.head()
#to get the start station names of the highest five
num=x2['count'].index.tolist()
#to get the number of the times used by start station
num1=x2['count'].tolist()
x2
```

```python
style.use('ggplot')
n=num
m=num1
plt.bar(n,m,color='#800080',align='center')
plt.xticks(rotation=45,ha='right')
plt.title('THE MOST FREQUENTLY USED START STATION')
plt.ylabel('COUNT')
plt.xlabel('START STATION')
plt.show() 
```
